```python
import pandas as pd
from datetime import datetime
import datetime as dt
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv('Daily_Balance_table.csv',sep=',')
#df['date_created-modified'] = df['date_created'].dt.date
df['date_created'] = pd.to_datetime(df['date_created'], format='%Y-%m-%d', errors='coerce')
df['date_created'] = df['date_created'].dt.strftime('%Y-%m-%d')
df = df.fillna(0)
df.head(5)
```

```python
df['date_created'] = df['date_created'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
x = df['date_created']
y = df['primary_account_balance']
# plot
plt.plot(x,y)
# beautify the x-labels
plt.gcf().autofmt_xdate()


plt.xlabel('data_created')
plt.ylabel('balance')
plt.title('End of Day Balance')
plt.savefig('End of Day Balance.png')
plt.show()
```
